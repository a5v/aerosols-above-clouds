```python
df0_ca = pd.read_csv("../create_atm/atmospheres/ca_0_1.csv", header = None)
df1_ca = pd.read_csv("../create_atm/atmospheres/ca_1_1.csv", header = None)
df2_ca = pd.read_csv("../create_atm/atmospheres/ca_2_1.csv", header = None)
df3_ca = pd.read_csv("../create_atm/atmospheres/ca_3_1.csv", header = None)
df4_ca = pd.read_csv("../create_atm/atmospheres/ca_4_1.csv", header = None)
df5_ca = pd.read_csv("../create_atm/atmospheres/ca_5_1.csv", header = None)
```

> load aerosol-cloud atmospheres

```python
df0_ac = pd.read_csv("../create_atm/atmospheres/ac_0_1.csv", header = None)
df1_ac = pd.read_csv("../create_atm/atmospheres/ac_1_1.csv", header = None)
df2_ac = pd.read_csv("../create_atm/atmospheres/ac_2_1.csv", header = None)
df3_ac = pd.read_csv("../create_atm/atmospheres/ac_3_1.csv", header = None)
df4_ac = pd.read_csv("../create_atm/atmospheres/ac_4_1.csv", header = None)
df5_ac = pd.read_csv("../create_atm/atmospheres/ac_5_1.csv", header = None)
```

## Impact of effective radius of aerosol

### wavelength = 0.56

```python
plt.plot(df0_ac[0], df0_ca[0])
plt.plot(df0_ac[1], df0_ca[1])
plt.plot(df0_ac[2], df0_ca[2])
plt.plot(df0_ac[3], df0_ca[3])
plt.xlabel('Reflection Function (a-c)')
plt.ylabel('Reflection Function (c-a)')
plt.title('effr_cloud = $0.1 \mu$m $\lambda$  = 0.56 $\mu$m')
plt.legend(['0.1','0.5','1','2'], title='effective radius of aerosol ($\mu$m)')
plt.show()
```

## Same effective radius of aerosol, same effective radius of cloud, different layering

- second looking at increasing effective radius of aerosol
