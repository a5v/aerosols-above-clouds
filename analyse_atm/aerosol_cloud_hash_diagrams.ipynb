# Notebook to look at impact of cloud effective radius (on aerosol-cloud atmosphere)

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Note optical thickness of clouds is fixed to 1

## Same effective radius of cloud, different effective radius of aerosol

First look at the files with effective radius of the cloud equal to 0.1

```python
df0 = pd.read_csv("../create_atm/atmospheres/ac_0_1.csv", header = None)
```

```python
df0
```

```python
plt.plot(df0[0], df0[4])
plt.plot(df0[1], df0[5])
plt.plot(df0[2], df0[6])
plt.plot(df0[3], df0[7])
plt.legend(['0.1','0.5','1','2'], title='effective radius of aerosol ($\mu$m)')
plt.xlabel('Reflection Function ($\lambda$ = 0.56 $\mu$m)')
plt.ylabel('Reflection Function ($\lambda$  = 1.59 $\mu$m)')
plt.title('effective radius of cloud = 0.1 $\mu$m')
plt.show()
```
