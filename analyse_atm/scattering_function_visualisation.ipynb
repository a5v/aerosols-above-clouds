# Plotting the normalised phase functions for different particulates

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre
from scipy.special import lpmv
import pandas as pd

import seaborn as sns
sns.set()
sns.set_style("whitegrid")
```

> defining range of scattering angle from 0 to 180 degrees

```python
N = 1000
xvals = np.linspace(0, np.pi, N)
```

## Rayleigh Scattering (of air molecules)

```python
leg_R = legendre(0) + 0.5*legendre(2)
```

```python
y_R = leg_R(np.cos(xvals))
```

```python
plt.plot(xvals, y_R)
plt.show
```

> on a polar plot

```python
thetas = np.linspace(0,2*np.pi,200)

plt.figure(figsize=(8,8))
r_R = lpmv(0, 0, np.cos(thetas)) + 0.5*lpmv(0, 2, np.cos(thetas))
plt.polar(thetas, abs(r_R))
plt.show
```

> polar plot
- note that the log scale is dodgy for the polar plot!!

```python
thetas = np.linspace(0,2*np.pi,200)
n_mom = 100

r1 = lpmv(0, 0, np.cos(thetas))

for i in range(1,n_mom): 
    r1 = r1 + (2*i+1)*df[0][i+3]*lpmv(0, i, np.cos(thetas))
    
r2 = lpmv(0, 0, np.cos(thetas))

for i in range(1,n_mom): 
    r2 = r2 + (2*i+1)*df[1][i+3]*lpmv(0, i, np.cos(thetas))
    
r3 = lpmv(0, 0, np.cos(thetas))

for i in range(1,n_mom): 
    r3 = r3 + (2*i+1)*df[2][i+3]*lpmv(0, i, np.cos(thetas))
    
r4 = lpmv(0, 0, np.cos(thetas))

for i in range(1,n_mom): 
    r4 = r4 + (2*i+1)*df[3][i+3]*lpmv(0, i, np.cos(thetas))
```

```python
plt.figure(figsize=(8,8))
plt.polar(thetas, abs(r1))
```

```python
plt.figure(figsize=(8,8))
plt.polar(thetas, abs(r2))
# plt.yscale('log')
```

```python
plt.figure(figsize=(8,8))
plt.polar(thetas, abs(r3))
```

```python
plt.figure(figsize=(8,8))
plt.polar(thetas, abs(r4))
```
